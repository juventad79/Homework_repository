```python
import pandas as pd
import scipy.stats as stats
import statistics as stat
import numpy as np
tit_df = pd.read_csv('data/train.csv')
tit_df.head()
```

```python
women = tit_df[tit_df['Sex'] == 'female']
women.describe()
```

```python
men = tit_df[tit_df['Sex'] == 'male']
men.describe()
```

```python
import pandas as pd
from scipy. stats import ttest_ind
stat, p = ttest_ind(men['Survived'], women['Survived'],equal_var= False)
print(p)
# p < 0.05
```
