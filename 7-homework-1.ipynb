```python
import math

def grad_f(x):
    return - 1 /(x*x) * math.cos(1/x)

print( grad_f(5) )
```

```python
import math 

def grad_f(x):
    return - 0.5 * math.exp(-x/2)

print( grad_f(2) )
```

Найти deltaf

```python
import numpy as np

def grad_f(x, y):
    grad_x = 2 * x + 2 * y
    grad_y = 2 * x + 2 * y - 3
    return np.array([grad_x, grad_y])

print( grad_f(2, 10) )
```

```python
import math 

def grad_f(x, y, z):
    grad_x = 2 * x + 2 * y - 2 * z
    grad_y = 2 * (x - z) - 4 * y - 6 * z
    grad_z = -2 * (x + y) -4 * z - 6 * y 
    return np.array([grad_x, grad_y,  grad_z])

print( grad_f(-1, -5, 0) )
```
