Детские достижения в увлечениях в диапазоне с 01.07.2014 по 01.01.2023

```python
import sqlite3
import pandas as pd
sql = sqlite3.connect('data/children_hobbies')
query = """
SELECT children.name as Имя, hobbies.name as Увлечение, achievement.date as Дата, achievement.name as Награда
FROM achievement
JOIN children_hobbies ON achievement.child_hobby_id = children_hobbies.id
JOIN children on children_hobbies.child_id = children.id
JOIN hobbies on children_hobbies.hobby_id = hobbies.id
where achievement.date >= date('2014-07-01') and achievement.date <= date('2023-01-01')
order by achievement.date
"""
pd.read_sql(query, sql)
```

Группировка полученных наград детьми(>=2)

```python
import sqlite3
import pandas as pd
sql = sqlite3.connect('data/children_hobbies')
query = """
SELECT children.name as 'Имя', hobbies.name as 'Увлечение', count(achievement.id) as 'Количество наград'
FROM achievement
JOIN children_hobbies ON achievement.child_hobby_id = children_hobbies.id
JOIN children on children_hobbies.child_id = children.id
JOIN hobbies on children_hobbies.hobby_id = hobbies.id
GROUP BY child_hobby_id
having count(achievement.child_hobby_id) >=2
"""
pd.read_sql(query, sql)
```

Информация кто занимается шахматами

```python
import sqlite3
import pandas as pd
sql = sqlite3.connect('data/children_hobbies')
query = """
SELECT *
FROM children_hobbies_period
JOIN children_hobbies ON children_hobbies_period.child_hobby_id = children_hobbies.id
JOIN children on children_hobbies.child_id = children.id
JOIN hobbies on children_hobbies.hobby_id = hobbies.id
where hobbies.name = 'Шахматы'
"""
pd.read_sql(query, sql)
```
