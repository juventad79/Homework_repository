Оценить с помощью доверительных интервалов:
Выживаемость всех пассажиров (оценить среднее по isSurvived) выживаемость женщин выживаемость людей с каютами первого класса



```python
import pandas as pd
import scipy.stats as stats
import statistics as stat
import numpy as np
tit_df = pd.read_csv('data/train.csv')
tit_df.head()
```

```python
tit_df.describe()
```

Доверительный интервал выживаемости всех пассажиров

```python
loc = tit_df["Survived"].mean()
std = tit_df["Survived"].std()
alpha = 0.95
stats.t.interval(1-alpha, len(tit_df)-1, loc = loc, scale = std/np.sqrt(len(tit_df)))
```

```python
women = tit_df[tit_df['Sex'] == 'female']
women.describe() 
```

Доверительный интервал выживаемости женщин

```python
   
alpha = 0.95
loc = women['Survived'].mean()
std = women['Survived'].std()

stats.t.interval(1-alpha, len(women)-1, loc = loc, scale = std/np.sqrt(len(women)))
```

Доверительный интервал выживаемости людей с каютами I класса

```python
class1 = tit_df[(tit_df['Pclass'] == 1)]
class1.describe()    

alpha = 0.95
loc = class1['Survived'].mean()
std =class1['Survived'].std()

stats.t.interval(1-alpha, len(class1)-1, loc = loc, scale = std/np.sqrt(len(class1)))
```
